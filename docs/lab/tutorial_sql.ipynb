# pyg-cell

pyg-cell is a package serving two purposes. 

* a thin layer over sqlalchemy with a few added tweaks to make it easier to use, more like mongodb
* a fully fledged document store, allowing the user to save arbitrary documents


```python
table = sql_table(db = 'test_db', schema = 'demo', table = 'demo', pk = dict(key = str, date = datetime.date), nullable = dict(binary = bin, number = float))
table
```

```python
from pyg import * 
import datetime
```

You notice the table is created automatically. There are three types of columns:
* pk: primary keys
* nullable: columns that can contain NULL



You can of course use the standard sqlalchemy objects...

```python
from sqlalchemy.types import NUMERIC, FLOAT, INT, Boolean, DECIMAL, VARBINARY, DATE, DATETIME, TIME, String, NVARCHAR
table = sql_table(db = 'test_db', schema = 'demo', table = 'demo', pk = dict(key = NVARCHAR(100), date = DATE), nullable = dict(binary = VARBINARY, number = NUMERIC))
```
